```python
from openai import OpenAI
```

```python
from getpass import getpass

api_key = getpass("Enter your OpenAI API Key: ")

client = OpenAI(api_key=api_key)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",
    input="What is an AI Agent?"
)

print(response.output_text)
```

```python
response = client.responses.create(
    model="gpt-4.1-mini",

    instructions="""
    You are an AI Engineering instructor.
    Explain concepts simply.
    Keep answers under 100 words.
    """,

    input="Explain MLOps."
)

print(response.output_text)
```

```python
def ask_agent(question):

    response = client.responses.create(
        model="gpt-4.1-mini",

        instructions="""
        You are an AI Engineering assistant.
        Explain clearly.
        Use simple language.
        Keep answers concise.
        """,

        input=question
    )

    return response.output_text
```

```python
print(ask_agent("What is Deep Learning?"))
```

```python
print(ask_agent("Explain TensorFlow Serving"))
```

```python
print(ask_agent("What is MLOps?"))
```

```python
def ask_stock_agent(question):

    response = client.responses.create(
        model="gpt-4.1-mini",

        instructions="""
        You are an experienced stock market analyst.
        Explain everything in simple language.
        """,

        input=question
    )

    return response.output_text

print(ask_stock_agent("Should investors diversify their portfolio?"))
```
